# 6.2

Calculating the average annual percentage return

```python
ret = []
for col in df:
    ret.append(df[col].pct_change(periods = 365).mean())
```

```python
AAPR = sum(ret)/len(ret)
AAPR
```

```python
vol = []
for col in df:
    vol.append(df[col].pct_change(periods = 365).std())
```

Volatility of the portfolio

```python
volatility = sum(vol)/len(vol)
volatility
```

```python
portfolio = pd.DataFrame()
portfolio['Returns'] = ret
portfolio['Volatility'] = vol
portfolio
```

importing the kMeans module to run divide the stocks into a set of clusters

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

Normalizing the data from each stock and their volatilities

```python
data = StandardScaler().fit_transform(portfolio)
data
```

Creating the weighted cluster sum of squares

```python
wcss=[]
for i in range(1,11): 
    kmeans = KMeans(n_clusters=i, init ='k-means++', max_iter=300, random_state=0)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
```

The elbow curve indicates the clusters should be set to 3

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
clustering_kmeans = KMeans(n_clusters=3)
model = clustering_kmeans.fit(data)
portfolio['cluster'] = clustering_kmeans.fit_predict(data)
portfolio
```

A scatter plot to represent the each stock in the cluster it belongs to

```python
fig , ax = plt.subplots(figsize=(12,12))

d = ax.scatter(portfolio['Returns'], portfolio['Volatility'], c=model.labels_.astype(float))
for i, txt in enumerate(df.columns):
    ax.annotate(txt, xy=(portfolio['Returns'][i], portfolio['Volatility'][i]))


plt.title('K-means Clustering for returns and volatility')
label = [1,2,3]
plt.legend(handles=d.legend_elements()[0], labels=label)

plt.xlabel("Returns")
plt.ylabel("Volatility")
plt.show()
```
