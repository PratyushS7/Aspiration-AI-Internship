## 1.4

In the following cells I extracted the Year and Month values from the original dataframe and appended these to the values originally contained in 'df'. This data will now be stored in df2

For calculating the Volume Weighted Average Price (VWAP), we need the following 4 columns of our original dataframe.

```python
dfc = df2[['Close Price','No. of Trades','Year','Month']]
dfc.head()
```

VWAP = sum(price * volume)/sum(volume)

```python
dfc[['Close Price']]=dfc[["Close Price"]].multiply(df["No. of Trades"], axis="index")
dfc.head()
```

Grouping the values having the same year and month together.

```python
td=dfc.groupby(['Year','Month']).sum()
td
```

These are the corresponding values of VWAP for each month

```python
td['VWAP'] = td['Close Price']/td['No. of Trades']
td['VWAP']
```

## 1.5

function to accept the parameter 'N' and return the average closing price for the last N days.

```python
def avg_price(N):
    sum = df[['Close Price']].tail(N).sum()
    sum = pd.Series(sum).values[0]
    return sum/N
```

This function is going to calculate the profit/loss percentange over the last 'N' days

```python
def profit_loss(N):
    l = df[['Close Price']].iloc[len(df.columns)]
    f = df[['Close Price']].iloc[len(df.columns)-N]
    val = ((l-f)/f)*100
    val = pd.Series(val).values[0]
    if(val<0):
        print('Loss= ',abs(val),'%')
    else:
        print('Profit= ',abs(val),'%')
```

```python
print('Average Price over last week=',avg_price(7))
profit_loss(7)

print('Average Price over last 2 weeks=',avg_price(14))
profit_loss(14)

print('Average Price over last month=',avg_price(30))
profit_loss(30)

print('Average Price over last 90 days=',avg_price(90))
profit_loss(90)

print('Average Price over last 180 days=',avg_price(180))
profit_loss(180)

print('Average Price over last year=',avg_price(365))
profit_loss(365)
```

## 1.6

Adding a column "Day_Perc_Change" which calculated the daily percentage change in the closing price of the stocks for consecutive days. 

We've made use of the pct_change() function is the pandas library.

```python
dfpc = df[['Close Price']].pct_change()*100
```

```python
dfpc["Close Price"].fillna(0,inplace=True)
dfpc.rename(columns={'Close Price':'Percentage_change'},inplace=True)

df = pd.merge(df, dfpc, left_index=True, right_index=True)
df.rename(columns={'Percentage_change':'Day_Perc_Change'},inplace=True)
```

## 1.7

Adding the 'Trend' column to the dataframe according to the given constraints on the percentage change.

```python
l = []
for i in df.Day_Perc_Change:
    if -0.5<= i <=0.5:
        l.append('Slight or no Change')
    elif 0.5<=i<=1:
        l.append('Slight Positive')
    elif -1<=i<=-0.5:
        l.append('Slight Negative')
    elif 1<=i<=3:
        l.append('Positive')
    elif -3<=i<=-1:
        l.append('Negative')
    elif 3<=i<=7:
        l.append('Among top gainers')
    elif -7<=i<=-3:
        l.append('Among top losers')
    elif i>7:
        l.append('Bull run')
    elif i<-7:
        l.append('Bear drop')
df['Trend']=l
```

## 1.8

Calculating the mean and the median of the 'Total traded Quanitity' column grouping them accoring the the 'Trends' we just set.

```python
dfmean = df.groupby('Trend').mean()
dfmean[['Total Traded Quantity']]
```

```python
dfmed = df.groupby('Trend').median()
dfmed[['Total Traded Quantity']]
```
